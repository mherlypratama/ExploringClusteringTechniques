# Data Understanding

```python
import pandas as pd
df = pd.read_csv('flight.csv')

df.sample(10)
```

```python
df.info()
```

Dataset ini merupakan data pelanggan program *Frequent Flyer*, yang berisi informasi tentang perilaku dan karakteristik pelanggan. Berikut adalah penjelasan tentang dataset Ini:

Dataset ini berisi 62988 baris data dan 20 kolom, yang dapat dikategorikan menjadi beberapa bagian:

Informasi Identitas

- MEMBER_NO-b: ID unik untuk setiap pelanggan
- GENDER: Jenis kelamin pelanggan
- AGE: Umur pelanggan
- WORK_CITY, WORK_PROVINCE, WORK_COUNTRY: Informasi tentang lokasi asal pelanggan

Informasi Penerbangan

- FIRST_FLIGHT_DATE: Tanggal penerbangan pertama pelanggan
- LAST_FLIGHT_DATE: Tanggal penerbangan terakhir pelanggan
- FLIGHT_COUNT: Jumlah penerbangan yang telah dilakukan pelanggan
- SEG_KM_SUM: Total jarak (km) penerbangan yang telah dilakukan pelanggan

Informasi Program Frequent Flyer

- FFP_DATE: Tanggal bergabung dengan program Frequent Flyer
- FFP_TIER: Tier pelanggan dalam program Frequent Flyer
- BP_SUM: Rencana perjalanan pelanggan
- SUM_YR_1, SUM_YR_2: Total kredit/point di tahun pertama dan kedua
- Points_Sum: Jumlah poin yang telah didapat pelanggan
- Point_NotFlight: Poin yang tidak digunakan oleh pelanggan

Informasi Lain

- LOAD_TIME: Tanggal data diambil
- AVG_INTERVAL: Rata-rata jarak waktu antara penerbangan
- MAX_INTERVAL: Maksimal jarak waktu antara penerbangan
- EXCHANGE_COUNT: Jumlah penukaran
- avg_discount: Rata-rata diskon yang didapat pelanggan

```python
# Handling kolom Gender dengan nilai modusnya
df['GENDER'].fillna(df['GENDER'].mode()[0], inplace=True)
```

```python
# Handling kolom lokasi dengan 'Unknown'
df[['WORK_CITY', 'WORK_PROVINCE']] = df[['WORK_CITY', 'WORK_PROVINCE']].fillna('Unknown')
```

```python
# Handling kolom country dengan nilai modusnya
df['WORK_COUNTRY'].fillna(df['WORK_COUNTRY'].mode().iloc[0], inplace=True)
```

```python
# Handling kolom AGE dengan nilai mediannya
df['AGE'].fillna(df['AGE'].median()[0], inplace=True)
```
